```python
# Konversi kolom "ammonia" ke float64
dataset['ammonia'] = pd.to_numeric(dataset['ammonia'], errors='coerce')
dataset['ammonia'].fillna(value=0, inplace=True)

# Konversi kolom "is_safe" ke float64
dataset['is_safe'] = pd.to_numeric(dataset['is_safe'], errors='coerce')

# Periksa kembali hasil konversi
print(dataset.info())
```

```python
dataset.describe()
```

```python
x = dataset.iloc[:,1:19].values
y = dataset.iloc[:,[-1]].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
print(x_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
# Cari indeks di mana terdapat nilai NaN pada y_train
nan_indices = np.argwhere(np.isnan(y_train))

# Hapus baris yang mengandung nilai NaN dari x_train dan y_train
x_train = np.delete(x_train, nan_indices, axis=0)
y_train = np.delete(y_train, nan_indices, axis=0)

classifer = KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=2)
classifer.fit(x_train,y_train)
```

```python
y_pred = classifer.predict(x_test)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

# Cari indeks di mana terdapat nilai NaN pada y_test
nan_indices = np.argwhere(np.isnan(y_test))

# Hapus baris yang mengandung nilai NaN dari x_test dan y_test
x_test = np.delete(x_test, nan_indices, axis=0)
y_test = np.delete(y_test, nan_indices, axis=0)

# Lakukan prediksi ulang setelah membersihkan data
y_prediksi = classifer.predict(x_test)

# Hitung matriks kebingungan setelah membersihkan data
cm = confusion_matrix(y_test, y_prediksi)
print(cm)
```
